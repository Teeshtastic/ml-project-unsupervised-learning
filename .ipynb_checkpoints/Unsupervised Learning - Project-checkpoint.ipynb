## Part II - KMeans Clustering


The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
# ELBOW METHOD
k_range = range(1, 10)
inertias = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled[spending_cols])
    inertias.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertias, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True, alpha=0.3)
plt.show()

# PERFORM CLUSTERING 
optimal_k = 3 
print(f"Using k = {optimal_k} clusters")

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled[spending_cols])

# ANALYZE CLUSTERS
print("\nCluster Sizes:")
cluster_sizes = df['Cluster'].value_counts().sort_index()
for cluster, size in cluster_sizes.items():
    print(f"Cluster {cluster}: {size} customers ({size/len(df)*100:.1f}%)")

print("\nCluster Characteristics (Average Spending):")
cluster_means = df.groupby('Cluster')[spending_cols].mean()
print(cluster_means.round(0))

# CLUSTER BY CHANNEL ANALYSIS
print("\nCluster vs Channel Analysis:")
channel_cluster = pd.crosstab(df['Cluster'], df['Channel'], normalize='index') * 100
print(channel_cluster.round(1))
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
# Create the hierarchical clustering
print("\nStep 2: Creating hierarchical clusters")
# This creates a linkage matrix showing how customers are grouped
linkage_matrix = linkage(df_scaled[spending_cols], method='ward')
print("Clustering linkage matrix created")

# Dendrogram
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, 
           truncate_mode='level', 
           p=10,  # Show only the last 10 merges
           leaf_rotation=90)

plt.title('Dendrogram - Hierarchical Clustering', fontsize=14)
plt.xlabel('Sample Index or Cluster Size')
plt.ylabel('Distance')

# Add lines to show potential cluster cuts
plt.axhline(y=10, color='red', linestyle='--', label='Cut for 3 clusters')
plt.axhline(y=8, color='blue', linestyle='--', label='Cut for 4 clusters')
plt.legend()
plt.show()

print("Dendrogram interpretation:")
print("- Tall vertical lines indicate good cluster separation")
print("- Red line creates 3 clusters")
print("- Blue line creates 4 clusters")

# Extract clusters
print("\nStep 4: Extracting clusters")
num_clusters = 3  

# Cut the dendrogram to get cluster assignments
clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')
df['Cluster'] = clusters

print(f"Created {num_clusters} clusters")

# Analyze clusters
print(f"\nStep 5: Analyzing clusters")

# Cluster sizes
for cluster_num in range(1, num_clusters + 1):
    count = (df['Cluster'] == cluster_num).sum()
    percentage = count / len(df) * 100
    print(f"Cluster {cluster_num}: {count} customers ({percentage:.1f}%)")

# Average spending by cluster
print(f"\nAverage spending by cluster:")
cluster_spending = df.groupby('Cluster')[spending_cols].mean()
print(cluster_spending.round(0))

# Channel distribution by cluster
print(f"\nChannel distribution by cluster:")
for cluster_num in range(1, num_clusters + 1):
    cluster_data = df[df['Cluster'] == cluster_num]
    channel1_pct = (cluster_data['Channel'] == 1).mean() * 100
    channel2_pct = (cluster_data['Channel'] == 2).mean() * 100
    print(f"Cluster {cluster_num}: {channel1_pct:.0f}% Channel 1, {channel2_pct:.0f}% Channel 2")

# Visualize results
print(f"\nStep 6: Visualizing clusters")
plt.figure(figsize=(15, 10))

# Create histogram for each spending category
for i, category in enumerate(spending_cols):
    plt.subplot(2, 3, i+1)
    
    # Plot histogram for each cluster
    for cluster_num in range(1, num_clusters + 1):
        cluster_data = df[df['Cluster'] == cluster_num][category]
        plt.hist(cluster_data, alpha=0.7, label=f'Cluster {cluster_num}', bins=15)
    
    plt.title(f'{category} Spending by Cluster')
    plt.xlabel('Spending Amount')
    plt.ylabel('Number of Customers')
    plt.legend()

plt.tight_layout()
plt.show()

# Summary
print("\n" + "="*50)
print("HIERARCHICAL CLUSTERING RESULTS")
print("="*50)
print("Customer cluster analysis:")

for cluster_num in range(1, num_clusters + 1):
    cluster_data = df[df['Cluster'] == cluster_num]
    avg_spending = cluster_data[spending_cols].mean()
    top_category = avg_spending.idxmax()
    dominant_channel = cluster_data['Channel'].mode()[0]
    
    print(f"\nCluster {cluster_num}:")
    print(f"   Size: {len(cluster_data)} customers")
    print(f"   Primary spending: {top_category}")
    print(f"   Dominant channel: Channel {dominant_channel}")
    
    if dominant_channel == 1:
        business_type = "Restaurants/Hotels"
    else:
        business_type = "Retail Stores"
    print(f"   Customer type: {business_type}")

print(f"\nThe algorithm identified {num_clusters} distinct customer segments.")
print("To try different cluster numbers, modify 'num_clusters' and re-run.")
```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
# Step 1: Data Preparation
print("STEP 1: DATA PREPARATION")
print("=" * 50)

spending_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Scale the data using robust scaling (recommended for this dataset with outliers)
scaler = RobustScaler()
df_scaled = df.copy()
df_scaled[spending_cols] = scaler.fit_transform(df[spending_cols])

print(f"Original data shape: {df.shape}")
print(f"Features for PCA: {spending_cols}")
print("Scaling method: Robust Scaling")

# Step 2: Perform PCA
print("\n\nSTEP 2: PERFORMING PCA")
print("=" * 50)

# Fit PCA with all components
pca = PCA()
pca_features = pca.fit_transform(df_scaled[spending_cols])

# Get explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

print("Explained Variance by Principal Component:")
for i, var in enumerate(explained_variance_ratio):
    print(f"PC{i+1}: {var:.4f} ({var*100:.2f}%)")

print(f"\nCumulative Explained Variance:")
for i, cum_var in enumerate(cumulative_variance):
    print(f"PC1 to PC{i+1}: {cum_var:.4f} ({cum_var*100:.2f}%)")

# Step 3: Analyze Principal Components
print("\n\nSTEP 3: PRINCIPAL COMPONENT ANALYSIS")
print("=" * 50)

# Component loadings (how much each original feature contributes to each PC)
component_loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(len(spending_cols))],
    index=spending_cols
)

print("Principal Component Loadings:")
print("(Shows how much each feature contributes to each component)")
print(component_loadings.round(4))

# Analyze first few components
print(f"\nAnalysis of Principal Components:")

for i in range(min(3, len(spending_cols))):  # Analyze first 3 PCs
    pc_name = f'PC{i+1}'
    print(f"\n{pc_name} (explains {explained_variance_ratio[i]*100:.1f}% of variance):")
    
    # Get loadings for this component
    loadings = component_loadings[pc_name].abs().sort_values(ascending=False)
    
    print(f"  Strongest contributors:")
    for feature, loading in loadings.head(3).items():
        direction = "positive" if component_loadings.loc[feature, pc_name] > 0 else "negative"
        print(f"    {feature}: {loading:.3f} ({direction})")

# Step 4: Visualizations
print("\n\nSTEP 4: VISUALIZATIONS")
print("=" * 50)

# Create comprehensive visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Principal Component Analysis Results', fontsize=16)

# 1. Scree Plot
axes[0, 0].bar(range(1, len(explained_variance_ratio)+1), explained_variance_ratio, 
               color='#3B82F6', alpha=0.7)
axes[0, 0].plot(range(1, len(cumulative_variance)+1), cumulative_variance, 
                'ro-', color='#EF4444', linewidth=2)
axes[0, 0].set_xlabel('Principal Component')
axes[0, 0].set_ylabel('Explained Variance Ratio')
axes[0, 0].set_title('Scree Plot')
axes[0, 0].axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='80% threshold')
axes[0, 0].legend()

# 2. Component Loadings Heatmap
sns.heatmap(component_loadings.iloc[:, :4], annot=True, cmap='RdBu_r', center=0, 
            ax=axes[0, 1], cbar_kws={'label': 'Loading'})
axes[0, 1].set_title('Component Loadings Heatmap')
axes[0, 1].set_ylabel('Original Features')

# 3. Biplot (PC1 vs PC2)
axes[0, 2].scatter(pca_features[:, 0], pca_features[:, 1], 
                   c=df['Channel'], cmap='viridis', alpha=0.6)
axes[0, 2].set_xlabel(f'PC1 ({explained_variance_ratio[0]*100:.1f}% variance)')
axes[0, 2].set_ylabel(f'PC2 ({explained_variance_ratio[1]*100:.1f}% variance)')
axes[0, 2].set_title('Biplot: PC1 vs PC2 (colored by Channel)')

# Add feature vectors to biplot
feature_vectors = pca.components_[:2].T
for i, feature in enumerate(spending_cols):
    axes[0, 2].arrow(0, 0, feature_vectors[i, 0]*3, feature_vectors[i, 1]*3,
                     head_width=0.1, head_length=0.1, fc='red', ec='red', alpha=0.7)
    axes[0, 2].text(feature_vectors[i, 0]*3.2, feature_vectors[i, 1]*3.2, feature,
                    fontsize=8, ha='center', va='center')

# 4. PC1 Distribution by Channel
for channel in [1, 2]:
    channel_data = pca_features[df['Channel'] == channel, 0]
    axes[1, 0].hist(channel_data, alpha=0.6, label=f'Channel {channel}', bins=20)
axes[1, 0].set_xlabel('PC1 Score')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('PC1 Distribution by Channel')
axes[1, 0].legend()

# 5. PC2 Distribution by Channel
for channel in [1, 2]:
    channel_data = pca_features[df['Channel'] == channel, 1]
    axes[1, 1].hist(channel_data, alpha=0.6, label=f'Channel {channel}', bins=20)
axes[1, 1].set_xlabel('PC2 Score')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('PC2 Distribution by Channel')
axes[1, 1].legend()

# 6. Cumulative Variance Plot
axes[1, 2].plot(range(1, len(cumulative_variance)+1), cumulative_variance, 
                'bo-', linewidth=2, markersize=8)
axes[1, 2].axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='80%')
axes[1, 2].axhline(y=0.95, color='orange', linestyle='--', alpha=0.7, label='95%')
axes[1, 2].set_xlabel('Number of Components')
axes[1, 2].set_ylabel('Cumulative Explained Variance')
axes[1, 2].set_title('Cumulative Explained Variance')
axes[1, 2].legend()

plt.tight_layout()
plt.show()

# Step 5: Dimensionality Reduction Recommendations
print("\n\nSTEP 5: DIMENSIONALITY REDUCTION RECOMMENDATIONS")
print("=" * 50)

# Find components needed for different variance thresholds
components_80 = np.argmax(cumulative_variance >= 0.8) + 1
components_95 = np.argmax(cumulative_variance >= 0.95) + 1

print(f"Components needed for 80% variance: {components_80}")
print(f"Components needed for 95% variance: {components_95}")

# Step 6: Create PCA-Transformed Dataset
print("\n\nSTEP 6: CREATING PCA-TRANSFORMED DATASET")
print("=" * 50)

# Create dataset with reduced dimensions
n_components = components_80  # Use 80% variance threshold
pca_reduced = PCA(n_components=n_components)
pca_features_reduced = pca_reduced.fit_transform(df_scaled[spending_cols])

# Create new dataframe with PCA features
df_pca = df.copy()
pca_column_names = [f'PC{i+1}' for i in range(n_components)]
for i, col_name in enumerate(pca_column_names):
    df_pca[col_name] = pca_features_reduced[:, i]

print(f"Created PCA-transformed dataset with {n_components} components")
print(f"New PCA features: {pca_column_names}")
print(f"Variance explained by selected components: {cumulative_variance[n_components-1]*100:.1f}%")

# Step 7: Interpretation and Insights
print("\n\nSTEP 7: INTERPRETATION AND INSIGHTS")
print("=" * 50)

# Analyze what each principal component represents
print("Principal Component Interpretation:")

# PC1 Analysis
pc1_loadings = component_loadings['PC1']
pc1_positive = pc1_loadings[pc1_loadings > 0].sort_values(ascending=False)
pc1_negative = pc1_loadings[pc1_loadings < 0].sort_values()

print(f"\nPC1 ({explained_variance_ratio[0]*100:.1f}% variance):")
if len(pc1_positive) > 0:
    print(f"  Positive direction: {', '.join(pc1_positive.head(3).index)}")
if len(pc1_negative) > 0:
    print(f"  Negative direction: {', '.join(pc1_negative.head(3).index)}")

# PC2 Analysis
pc2_loadings = component_loadings['PC2']
pc2_positive = pc2_loadings[pc2_loadings > 0].sort_values(ascending=False)
pc2_negative = pc2_loadings[pc2_loadings < 0].sort_values()

print(f"\nPC2 ({explained_variance_ratio[1]*100:.1f}% variance):")
if len(pc2_positive) > 0:
    print(f"  Positive direction: {', '.join(pc2_positive.head(3).index)}")
if len(pc2_negative) > 0:
    print(f"  Negative direction: {', '.join(pc2_negative.head(3).index)}")

# Business interpretation
print(f"\nBusiness Interpretation:")
print("Based on the component loadings and channel analysis:")

# Analyze PC1 vs Channel relationship
pc1_channel1_mean = pca_features[df['Channel'] == 1, 0].mean()
pc1_channel2_mean = pca_features[df['Channel'] == 2, 0].mean()

if pc1_channel1_mean > pc1_channel2_mean:
    print(f"- PC1 likely represents Restaurant/Hotel vs Retail distinction")
    print(f"- Higher PC1 scores = Restaurant/Hotel customers (Channel 1)")
    print(f"- Lower PC1 scores = Retail customers (Channel 2)")
else:
    print(f"- PC1 likely represents Retail vs Restaurant/Hotel distinction")
    print(f"- Higher PC1 scores = Retail customers (Channel 2)")
    print(f"- Lower PC1 scores = Restaurant/Hotel customers (Channel 1)")

print(f"\n- PC2 likely captures spending scale or specific product preferences")
print(f"- Together, PC1 and PC2 explain {(explained_variance_ratio[0] + explained_variance_ratio[1])*100:.1f}% of variance")

# Summary statistics
print(f"\n\nSUMMARY STATISTICS")
print("=" * 50)
print(f"Original dimensions: {len(spending_cols)}")
print(f"Recommended reduced dimensions: {n_components}")
print(f"Variance retained: {cumulative_variance[n_components-1]*100:.1f}%")
print(f"Variance lost: {(1-cumulative_variance[n_components-1])*100:.1f}%")
print(f"Dimensionality reduction: {((len(spending_cols)-n_components)/len(spending_cols))*100:.1f}%")

print(f"\nPCA reveals the underlying structure of customer spending patterns.")
print(f"Use df_pca with PC features for downstream machine learning tasks.")
```

Geographic Concentration with Growth Opportunities: Region 3 dominates the customer base with 72% of all customers (316 out of 440), while Regions 1 and 2 represent significant untapped markets with only 77 and 47 customers respectively. This concentration suggests either strong market penetration in one area or potential expansion opportunities in underserved regions.
Principal Component Analysis Confirms Business Type as Primary Driver: PCA reveals that the first principal component explains the largest variance in the data and clearly separates customers along the restaurant/hotel versus retail store dimension. The underlying structure shows that combinations of Fresh vs Grocery/Milk/Detergents spending patterns are the most important factors for understanding customer behavior, confirming that business type is the fundamental driver of purchasing decisions.
